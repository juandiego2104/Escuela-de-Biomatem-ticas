# Solucion de Ecuaciones Diferenciales

### Modelo SIR

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#parametros del modelo
beta, gama=0.4,0.7
tmax=365
t=np.linspace(0,tmax,10*tmax)
```

```python
#definicion del lado derecho del modelo
def modSIR(y,t,beta,gama):
    S,I,R = y
    dSdt =-beta*S*I
    dIdt =beta*S*I-gama*I
    dRdt =gama*I
    return dSdt, dIdt, dRdt
```

```python

#Condiciones iniciles
S0,I0,R0=25,15,0
y0=S0,I0,R0
#
solucion=odeint(modSIR,y0,t,args=(beta,gama))
S,I,R=solucion.T
```

```python
solucion.shape
```

```python
#grafica de la solucion
%matplotlib inline
fig=plt.figure()
plt.plot(t,S,'b',label="Suceptibles")
plt.plot(t,I,'r',label="Infecciosos")
plt.plot(t,R,'g',label="Recuperados")
plt.xlabel("Tiempo")
plt.ylabel("Numero")
plt.xlim(0,10)
legend=fig.legend()
legend.get_frame().set_alpha(0.5)

```
